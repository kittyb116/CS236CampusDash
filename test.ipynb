```python
def read_dashers(fname):
    # Open the file
    file = open(fname, "r")
    # Set up your list of dashers
    dashers=[]
    #id = 1
    # Next, read the agents and build the list
    file.readline()
    for line in file:
        x,y,z = line.split(",")
        dashers.append((int(x),int(y),int(z)))
    # Close the file safely after done reading
    file.close()
    return dashers
```

```python
read_dashers("input/project files/dashers.csv")
```
